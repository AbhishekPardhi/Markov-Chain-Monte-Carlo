```julia
using Plots
function DIT(λ)
    U = rand(Float64)
    i = 0
    p = exp(-λ)
    A = p
    X = 0
    while true
        if U < A
            X = i
            break
        else
            p = λ/(i + 1)*p
            A = A + p
            i = i + 1 
        end
    end
    return X
end
```

The corresponding Plots for the discrete Random Variable can be found below as 10000 iid's.

```julia
λ = 10
rep = 10000
X = [DIT(λ) for i in 1:rep]
histogram(1:rep, X, title = "Discrete Inverse Transform sampler\nfor Poisson distribution(λ=10)", label = "Random Variable")
```

```julia
λ = 25
rep = 10000
X = [DIT(λ) for i in 1:rep]
histogram(1:rep, X, title = "Discrete Inverse Transform sampler\nfor Poisson distribution(λ=25)", label = "Random Variable")
```

```julia
λ = 100
rep = 10000
X = [DIT(λ) for i in 1:rep]
histogram(1:rep, X, title = "Discrete Inverse Transform sampler\nfor Poisson distribution(λ=100)", label = "Random Variable")
```

Hence, it can be seen that the mean(expected value) of the random variable of the Discrete Inverse Transform sampler for Poisson distribution is equal to λ.
