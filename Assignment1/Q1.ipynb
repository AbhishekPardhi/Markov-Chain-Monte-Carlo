```julia
using Plots
function DIT(λ)
    U = rand(Float64)
    i = 0
    p = exp(-λ)
    A = p
    X = 0
    while true
        if U < A
            X = i
            break
        else
            p = λ/(i + 1)*p
            A = A + p
            i = i + 1 
        end
    end
    return X
end
```

The corresponding Plot for the discrete Random Variable can be found below as 10000 iid's.

```julia
λ = 1
rep = 10000
X = [DIT(λ) for i in 1:rep]
plot(1:rep, X, title = "Discrete Inverse Transform sampler\nfor Poisson distribution")
```
